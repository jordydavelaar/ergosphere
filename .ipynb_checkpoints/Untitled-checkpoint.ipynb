```python
from params import Params
from universe import AssetUniverse
from data_loader import DataLoader
from backtester import AdaptiveBacktester, BuyAndHoldBacktester
from strategy import MomentumStrategy
import matplotlib.pyplot as plt

# Step 1: Set params and universe
params = Params()
params.load_overrides_from_file()
universe = AssetUniverse(params=params)

# Step 2: Load data
data = DataLoader(params=params, universe=universe)
data.download_data()

# Step 3: Create strategies
momentum_strategy = MomentumStrategy(data=data, params=params, universe=universe)
momentum_strategy.generate_signal()

# Step 4: Run backtests
adaptive = AdaptiveBacktester(data=data, strategy=momentum_strategy, params=params)
adaptive.run()

buy_and_hold = BuyAndHoldBacktester(data=data, params=params)
buy_and_hold.run()

# Optional: Show performance summary
from pandas import DataFrame
DataFrame(adaptive.performance_log + buy_and_hold.performance_log)
```

```python
import numpy as np
from datetime import datetime
from universe    import AssetUniverse
from params      import Params
from data_loader import DataLoader
from strategy    import MomentumStrategy
from backtester  import AdaptiveBacktester
from surrogate_model import SurrogateModel

# Setup parameters
params = Params(
    # ── Universe sizing ────────────────────────────────────
    top_n                   = 20,        # number of cryptos to consider
    start_date              = "2023-05-01",
    end_date                = "2025-03-01",
    rebalance_interval      = 3,
    
    # ── Capital & Timing ───────────────────────────────────
    initial_capital         = 10_000.0,  # starting USD
    time_step               = 7,         # rebalance every 4 weeks
    lookback                = 10,        # signal vs realized return horizon
    
    # ── Risk & Weights ─────────────────────────────────────
    risk_aversion           = 0.0,       # how much you penalize variance
    regularization_param    = 1e-6,      # tiny ridge‐like shrinkage
    rebalance_param         = 1e-4,      # penalty on large weight shifts
    L1_reg                  = 1e-5,      # adds slight sparsity

    max_weight_shift        = 0.1,      # max 10% change per rebalance
    max_weight              = 0.7,       # cap any single coin at 50%

    # ── Transaction Costs & Fees ──────────────────────────
    transaction_cost        = 0.001,     # 0.1% per trade
    bid_ask_spread          = 0.002,     # 0.2% round‐trip slippage
    tax_rate                = 0.20,      # 20% on realized gains
    management_fee          = 0.005,     # 0.5% per annum
    inflation_rate          = 0.02,      # 2% per annum
    leverage_interest       = 0.08,      # 8% borrowing cost

    # ── Signal Horizons ───────────────────────────────────
    horizons                = [15, 30, 40],      # medium‐term
    horizon_weights         = [1.0, 0.8, 0.5,0.25],
    short_horizons          = [10, 15,20],           # for high‐vol periods
    short_horizon_weights   = [1.0, 0.6,0.4],
    high_vol_zscore_threshold = 2.0,    # above 1σ triggers short horizons
    volatility_cash_boost     = False,   # lean into cash when vol is high
    volatility_window         = 14,     # 2‐week vol rolling

    ma_slow_window           = 20,      # for death‐cross checks
    ma_fast_window           = 5,

    # ── Bull / Bear Regime Params ─────────────────────────
    bull_weight_lower       = 0.0,      # long‐only
    bull_weight_upper       = 0.6,
    bull_net_target         = 1.0,
    bull_leverage_limit     = 1.0,

    allow_shorting        = False,
    bear_weight_lower     = -0.3,    # allow up to 50% short
    bear_weight_upper     = 0.3,
    bear_net_target       = -0.1,    # net short 30%
    bear_leverage_limit   = 1.5,     # allow gross ≤150%

    # ── Asset Class Ranges ─────────────────────────────────
    crypto_weight_range     = (0.0, 1.0),
    bond_weight_range       = (0.0, 0.0),  # we’re crypto‐only here
    cash_weight_range       = (0.0, 0.0),

    # ── Bear-Market Detector ────────────────────────────────
    drawdown_thresh         = -0.15,     # 15% drawdown triggers flag
    drawdown_window         = 60,        # 3‐month lookback
    vol_z_thresh            = 1.0,       # 1σ move above mean vol
    ma_below_ratio          = 0.3,       # 30% of days below MA = bearish
    ma_below_days           = 20,
    vol_window              = 30        # 1‐month vol lookback
)



# Initialize universe and load price data
print("Initializing universe and loading data...")
universe = AssetUniverse(params)
data = DataLoader(universe, params)
data.download_data(overwrite=True)  # Now holds data.prices, data.CVI, data.VIX

# Initialize strategy and generate signals
print("Generating momentum signals...")
strategy = MomentumStrategy(universe, params, data)
strategy.generate_signal()


# Custom parameter specifications for grid
param_specs = [
    ("max_weight_shift", np.array([0.1,0.5,1.0])),
    ("bull_weight_upper", np.linspace(0.1, 1.0, 3)),
]

# Initialize Surrogate Model
model = SurrogateModel(params, universe, data, param_specs=param_specs)

# Define grid
model.define_grid()  # Add this call if missing


# Run backtests with parallel processing
model.run_backtests(n_processes=8)

# Fit surrogate models
model.fit_surrogate_models()

# Optimize using RBF
opt_result = model.optimize_rbf()
print(f"Optimal CAGR via RBF: {-opt_result.fun:.4f}, at {opt_result.x}")

# GP sampling
x_query, y_pred, y_std = model.predict_gp(n_samples=1000)
print(f"GP best predicted CAGR: {np.max(y_pred):.4f}")

# Save the model
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
model.save(f"surrogate_model_{timestamp}.pkl", opt_result=opt_result)
```
